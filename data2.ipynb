```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import os
print(os.getcwd())
```

```python
data = pd.read_csv(r"C:\Users\PC\Documents\prodigy\dataset2.csv")
```

```python
print(data.isnull().sum())
```

```python
print(data.isnull().sum())
```

```python
print(data.describe())
```

```python
#Survival Rate
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.histplot(data['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

```

```python
The above distrubution shows high number of young passengers.
```

```python
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.title('Survival Distribution by Class')
plt.show()
```

```python
#pair plot
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
sns.pairplot(data, hue='Survived', vars=['Age', 'Fare', 'FamilySize'])
plt.title('Pair Plot')
plt.show()
```

```python
data['IsAlone'] = (data['FamilySize'] == 1).astype(int)

# Survival rate by IsAlone
sns.countplot(x='IsAlone', hue='Survived', data=data)
plt.title('Survival Rate by IsAlone')
plt.show()

```

```python
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1

# Survival rate by family size
sns.countplot(x='FamilySize', hue='Survived', data=data)
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
Findings:
#Women had a higher survival rate compared to men.
#Passengers in first class had a higher survival rate compared to those in second and third class.
#The age distribution shows a higher number of younger passengers.
#Passengers with smaller family sizes had a higher survival rate.
#Passengers who were alone had a lower survival rate.
```
